```python
from pympler import asizeof

def print_memory(obj, name):
    size = asizeof.asizeof(obj)
    print(f"{name}: {size / 1024**2:.2f} MB")
```

```python
import joblib
from pathlib import Path

# Change this path if needed
COLLABRATIVE_DIR = Path("C:/Users/dell/OneDrive - Government Polytechnic, Ahmedabad/Internship TRT/Work/Movie_Recommandation/model_supporting_file/Collabrative_Based_Data")

# Load artifacts
movie_dict = joblib.load(COLLABRATIVE_DIR / "movie_dict.pkl")
all_movies_id = list(joblib.load(COLLABRATIVE_DIR / "all_movies_id.pkl"))

print("=" * 80)
print("BASIC INFORMATION")
print("=" * 80)

print(f"movie_dict type      : {type(movie_dict)}")
print(f"all_movies_id type   : {type(all_movies_id)}")
print(f"movie_dict length    : {len(movie_dict)}")
print(f"all_movies_id length : {len(all_movies_id)}")

print("\n")

print("=" * 80)
print("FIRST DICTIONARY ENTRY")
print("=" * 80)

first_key = next(iter(movie_dict.keys()))

print(f"First Key        : {first_key}")
print(f"Key Type         : {type(first_key)}")
print(f"First Value      : {movie_dict[first_key]}")
print(f"Value Type       : {type(movie_dict[first_key])}")

print("\n")

print("=" * 80)
print("FIRST 10 MOVIE IDS")
print("=" * 80)

print(all_movies_id[:10])

print("\nMovie ID Type:")
print(type(all_movies_id[0]))

print("\n")

print("=" * 80)
print("CHECK KNOWN MOVIE (318)")
print("=" * 80)

print("Lookup using int key")
print("--------------------")
print("318 in dict:", 318 in movie_dict)
print("Value:", movie_dict.get(318))

print()

print("Lookup using string key")
print("-----------------------")
print("'318' in dict:", "318" in movie_dict)
print("Value:", movie_dict.get("318"))

print("\n")

print("=" * 80)
print("CHECK MOVIE FROM YOUR API OUTPUT")
print("=" * 80)

movie_id = all_movies_id[0]

print(f"Movie ID: {movie_id}")
print(f"Movie ID Type: {type(movie_id)}")

print()

print("Lookup using original type")
print("--------------------------")
print(movie_id in movie_dict)
print(movie_dict.get(movie_id))

print()

print("Lookup using int()")
print("------------------")
print(int(movie_id) in movie_dict)
print(movie_dict.get(int(movie_id)))

print()

print("Lookup using str()")
print("------------------")
print(str(movie_id) in movie_dict)
print(movie_dict.get(str(movie_id)))

print("\n")

print("=" * 80)
print("FIRST 10 LOOKUPS")
print("=" * 80)

for mid in all_movies_id[:10]:
    title = (
        movie_dict.get(mid)
        or movie_dict.get(int(mid), None)
        or movie_dict.get(str(mid), None)
    )

    print(f"{mid:<10} --> {title}")
```
